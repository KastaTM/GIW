```python
#OPCIÓN 1
def opcion1(headers):
    
    url = "https://vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com/api/covid-ovid-data/"
        
    response = requests.request("GET", url, headers=headers)
    informacion = response.json()

    option = "S"
    while(option =="S"):
        funOpcion1(informacion)
        option = volver_a_buscar()

def funOpcion1(info):
    lista_continentes = []
    lista_paises = []
    for i in info:
        if i["Continent"] != None:
            if i["Continent"] not in lista_continentes:
                lista_continentes.append(i["Continent"])
                
    print("Seleccione el continente del que desea obtener las estadísticas: (Número)")
    for i in range(len(lista_continentes)-1):
        print(i+1,lista_continentes[i])
    opCont = int(input())
    
    while(opCont<0 and opCont > len(lista_continentes)+1):
        print("Seleccione una opción valida")
        opCont = int(input())
    
    for i in info:
        if i["Continent"] == lista_continentes[opCont-1]:
            if i["Country"] not in lista_paises:
                lista_paises.append(i["Country"])
            
    print("Seleccione el país del que desea obtener las estadísticas: (Número)")
    for i in range(len(lista_paises)):
        print(i+1,lista_paises[i])
    opPais = int(input())
    
    while(opPais<0 and opPais > len(lista_paises)+1):
        print("Seleccione una opción valida")
        opPais = int(input())
        
    print("Las estadísticas de",lista_paises[opPais-1],lista_continentes[opCont-1],"son:")
    
    casos_total = 0
    nuevos_casos = 0
    muertes_totales = 0
    nuevas_muertes = 0
    test_totales = 0
    for i in info:
        if i["Country"] == lista_paises[opPais-1]:
            casos_total+=i['total_cases']
            nuevos_casos+=i['new_cases']
            muertes_totales+=i['total_deaths']
            nuevas_muertes+=i['new_deaths']
            test_totales+=i['total_tests']
        
    print("Casos totales:",casos_total)
    print("Nuevo casos:",nuevos_casos)
    print("Muertes totales:",muertes_totales)
    print("Nuevas muertes:",nuevas_muertes)
    print("Test totales:",test_totales)

    
    
```
